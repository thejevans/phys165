```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.rcParams['figure.dpi'] = 300

def gaussian(x, A, mu, sigma): return A*np.exp(-(x - mu)**2/(2*sigma**2))

# 100 trials of 10000 tosses. 1 = heads, 0 = tails
tosses = np.random.randint(2, size = (100,10000))

# sum up all heads in each trial
sums = np.sum(tosses, axis = 0)

xs = np.linspace(30,70,100)

plt.hist(sums, bins = range(101), histtype = 'step', align = 'left')
plt.plot(xs, gaussian(xs, 800, 50, 5))
plt.xlim([30,70])
plt.title('Coin toss distribution')
plt.xlabel('# of heads')
plt.ylabel('counts')
plt.show()
```
