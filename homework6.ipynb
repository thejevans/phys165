```python
import numpy as np
from matplotlib import rcParams
import matplotlib.pyplot as plt

# make plots look nice in jupyter notebook
rcParams['figure.dpi'] = 300

# import csv data
time, amp = np.loadtxt('./Homework6_data_packed.csv', delimiter=',', unpack = True)

# compute the fourier transform of the waveform
y = np.fft.rfft(amp)

# make an array of frequencies to match with the output of np.fft.rfft
# format is (1/total_time) * range(length_of_rfft_output)
f = np.arange(len(y))/time[-1]

# format plotting area
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

# plot the waveform
ax = axes[0]
ax.plot(time, amp)
ax.set_xlabel('time (s)')
ax.set_ylabel('amplitude')
ax.set_title('Amplitude vs. Time')

# plot the magnitude of the rfft output vs frequency
ax = axes[1]
ax.plot(f, np.abs(y))
ax.set_xlabel('frequency (Hz)')
ax.set_ylabel('fourier transform of waveform')
ax.set_title('Fourier Transform vs. Frequency')

plt.show()
```
