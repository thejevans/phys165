```python
'''
John Evans
9/18/2018
'''
import numpy as np
import matplotlib.pyplot as plt

# magic command for printing plots inline in a jupyter notebook
%matplotlib inline

# acceleration due to gravity
g = 9.80665

def projectile_motion(v_0, theta, steps=100):
    '''
    v_0:   initial velocity in meters/second
    theta: intial angle initial angle from the horizontal in degrees
    steps: number of steps to divide output into
    '''
    # x and y components of initial velocity
    v_0x = v_0 * np.cos(np.radians(theta))
    v_0y = v_0 * np.sin(np.radians(theta))

    # time of impact from y velocity component
    # v_f  = v_0y + a * t_f
    # t_f  = (v_fy - v_0y) / a
    # v_fy = -v_0y
    # t_f  = -2 * v_0y / a
    # a    = -g
    t_f  = 2 * v_0y / g

    # get timestep values from kinematic equations
    t = np.linspace(0,t_f,steps)
    x = v_0x * t
    y = v_0y * t - 0.5 * g * t**2
    
    return x,y,t

# resize plot window
plt.figure(figsize=(15, 10))
plt.subplots_adjust(hspace=.3)

# V0 = 75.0 m/s and  = 60 deg
x,y,t = projectile_motion(75, 60)

# first subplot: single plot with both y(t) and x(t) vs t
plt.subplot(221)
plt.plot(t, x, label='x(t)')
plt.plot(t, y, label='y(t)')
plt.legend()
plt.xlabel('t [s]')
plt.ylabel('position [m]')
plt.title(r'single plot with both y(t) and x(t) vs t, $v_0$ = 75 m/s and $\theta=60^{\circ}$')

# second subplot: y(t) vs x(t)
plt.subplot(222)
plt.plot(x, y)
plt.xlabel('x(t) [m]')
plt.ylabel('y(t) [m]')
plt.title(r'y(t) vs x(t), $v_0$ = 75 m/s and $\theta=60^{\circ}$')

# V0 = 75.0 m/s
# third subplot: single plot y(t) vs x(t) with four curves on it for  = 15, 30, 45, 60 deg
thetas = [15, 30, 45, 60]
data   = [projectile_motion(75, theta) for theta in thetas]
plt.subplot(223)
for (x, y, _), theta in zip(data,thetas):
    plt.plot(x, y, label=r'$\theta={:g}^{{\circ}}$'.format(theta))
plt.legend()
plt.xlabel('x(t) [m]')
plt.ylabel('y(t) [m]')
plt.title(r'single plot y(t) vs x(t) with four curves on it, $v_0$ = 75 m/s')

#  = 60 deg
# fourth subplot: single plot y(t) vs x(t) with three curves on it for V0 = 25, 50, 75 m/s
v_0s = [25, 50, 75]
data = [projectile_motion(v_0, 60) for v_0 in v_0s]
plt.subplot(224)
for (x, y, _), v_0 in zip(data, v_0s):
    plt.plot(x, y, label=r'$v_0={:g}$ m/s'.format(v_0))
plt.legend()
plt.xlabel('x(t) [m]')
plt.ylabel('y(t) [m]')
plt.title(r'single plot y(t) vs x(t) with three curves on it, $\theta=60^{\circ}$')

# display plots
plt.show()
```
