```python
'''
John Evans
9/13/2018




'''
import numpy as np
import pandas as pd

# acceleration due to gravity
g = 9.80665

v_0   = float(input('initial velocity magnitude (m/s):'))
theta = float(input('initial angle from the horizontal (degrees):'))

# x and y components of initial velocity
v_0x = v_0 * np.cos(np.radians(theta))
v_0y = v_0 * np.sin(np.radians(theta))

# time of impact from y velocity component
# v_f  = v_0y + a * t_f
# t_f  = (v_fy - v_0y) / a
# v_fy = -v_0y
# t_f  = -2 * v_0y / a
# a    = -g

t_f  = 2 * v_0y / g

# final x position from x velocity component
# x_f = x_0 + v_0x * t_f
# x_0 = 0

x_f = v_0x * t_f

print('\nFinal x position:\t{:.4f} meters\nTime of impact:\t\t{:.4f} seconds\n'.format(x_f, t_f))

# Chosen v_0: 10 m/s
# 15 degree output: Final x position:	5.0986 meters
#                   Time of impact:		0.5278 seconds
# 30 degree output: Final x position:	8.8310 meters
#                   Time of impact:		1.0197 seconds
# 45 degree output: Final x position:	10.1972 meters
#                   Time of impact:		1.4421 seconds
# 60 degree output: Final x position:	8.8310 meters
#                   Time of impact:		1.7662 seconds

# get timestep values from kinematic equations
t = np.linspace(0,t_f,20)
x = v_0x * t
y = v_0y * t - 0.5 * g * t**2

# build dataframe for output
df = pd.DataFrame(np.array([t, x, y]).T, columns=['time (s)', 'x position (m)', 'y position (m)'])
pd.options.display.float_format = '{:.4f}'.format
print(df)
```
